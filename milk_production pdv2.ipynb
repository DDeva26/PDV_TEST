```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('milk_production.csv')
print(df)
```

```python
#1
df["maxmilk2013_14"] = df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"]

df[df["maxmilk2013_14"]==df["maxmilk2013_14"].max()]["State/ UT Name"]
```

```python
#2
df["2010"]=df["Cow Milk-2010-11"] + df["Boffalo Milk-2010-11"] + df["Goat Milk-2010-11"]
df["2011"]=df["Cow Milk-2011-12"] + df["Boffalo Milk-2011-12"] + df["Goat Milk-2011-12"]
df["2013"]=df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"]
df["2014"]=df["Cow Milk-2014-15"] + df["Boffalo Milk-2014-15"] + df["Goat Milk-2014-15"]
df["2015"]=df["Cow Milk-2015-16"] + df["Boffalo Milk-2015-16"] + df["Goat Milk-2015-16"]

print("Top 5 cities of 2010")
print(df.sort_values("2010",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2011")
print(df.sort_values("2011",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2013")
print(df.sort_values("2013",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2014")
print(df.sort_values("2014",ascending=False)["State/ UT Name"].head())
print("Top 5 cities of 2015")
print(df.sort_values("2015",ascending=False)["State/ UT Name"].head())


```
